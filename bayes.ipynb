P(B,D,T)

```python
bdt = df.loc[(df['PTS'] > df['PTS'].mean()) & (df['2P%'] > 0.55) & (df['3P%'] > 0.38)].shape[0]/df['PTS'].shape[0]
print('bdt: ', bdt)
bd_t = df.loc[(df['PTS'] > df['PTS'].mean()) & (df['2P%'] > 0.55) & (df['3P%'] <= 0.38)].shape[0]/df['PTS'].shape[0]
print('bd_t: ', bd_t)
b_dt = df.loc[(df['PTS'] > df['PTS'].mean()) & (df['2P%'] <= 0.55) & (df['3P%'] > 0.38)].shape[0]/df['PTS'].shape[0]
print('b_dt: ', b_dt)
b_d_t = df.loc[(df['PTS'] > df['PTS'].mean()) & (df['2P%'] <= 0.55) & (df['3P%'] <= 0.38)].shape[0]/df['PTS'].shape[0]
print('b_d_t: ', b_d_t)
_bdt = df.loc[(df['PTS'] <= df['PTS'].mean()) & (df['2P%'] > 0.55) & (df['3P%'] > 0.38)].shape[0]/df['PTS'].shape[0]
print('_bdt: ', _bdt)
_bd_t = df.loc[(df['PTS'] <= df['PTS'].mean()) & (df['2P%'] > 0.55) & (df['3P%'] <= 0.38)].shape[0]/df['PTS'].shape[0]
print('_bd_t: ', _bd_t)
_b_dt = df.loc[(df['PTS'] <= df['PTS'].mean()) & (df['2P%'] <= 0.55) & (df['3P%'] > 0.38)].shape[0]/df['PTS'].shape[0]
print('_b_dt: ', _b_dt)
_b_d_t = df.loc[(df['PTS'] <= df['PTS'].mean()) & (df['2P%'] <= 0.55) & (df['3P%'] <= 0.38)].shape[0]/df['PTS'].shape[0]
print('_b_d_t: ', _b_d_t)
```

```python
dt = df.loc[(df['2P%'] > 0.55) & (df['3P%'] > 0.38)].shape[0]/df['3P%'].shape[0]
print('dt: ', dt)
d_t = df.loc[(df['2P%'] > 0.55) & (df['3P%'] <= 0.38)].shape[0]/df['3P%'].shape[0]
print('d_t: ', d_t)
_dt = df.loc[(df['2P%'] <= 0.55) & (df['3P%'] > 0.38)].shape[0]/df['3P%'].shape[0]
print('_dt: ', _dt)
_d_t = df.loc[(df['2P%'] <= 0.55) & (df['3P%'] <= 0.38)].shape[0]/df['3P%'].shape[0]
print('_d_t: ', _d_t)
```
